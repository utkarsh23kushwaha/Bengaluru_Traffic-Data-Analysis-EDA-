```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Traffic_Offence_Data(2020-22).csv")

#comverting all the entires to numeric type and replacing NaN values with 0
data[['2020','2021','2022']]=data[['2020','2021','2022']].apply(pd.to_numeric, errors='coerce').fillna(0)

#adding columns for percentage occurence of each type of offence for every year
data['2020_%']=data['2020']*100/ data['2020'].sum()
data['2021_%']=data['2021']*100/ data['2020'].sum()
data['2022_%']=data['2022']*100/ data['2020'].sum()
data 

```

```python
D_20=data[['S.no','Type of Offences','2020','2020_%']].sort_values(by='2020',ascending=False).head(10)
D_20
```

```python

D_21=data[['S.no','Type of Offences','2021','2021_%']].sort_values(by='2021',ascending=False).head(10)
D_21
```

```python
D_22=data[['S.no','Type of Offences','2022','2022_%']].sort_values(by='2022',ascending=False).head(10)
D_22
```

```python

data.plot(x="Type of Offences", y=["2022","2021","2022"], kind="line",figsize=(10,8))
plt.xticks(rotation=90)
```

```python
data2=pd.read_csv("Traffic_Accidents_Data(2020-22).csv")

#dropping the rows with zonal total figures
data2_1=data2.drop([21,40,46,47])
data2_1

```

```python
#Location vs Number of Accidents in 2020 (Most 5)
d1=data2_1[['Total_Cases_20','Station_Name']].sort_values(by='Total_Cases_20',ascending=False).head(5)
d1.plot(x='Station_Name',y='Total_Cases_20',figsize=(5,4),kind='bar',legend=None, color='#FF5D00')
plt.xticks(rotation=90)
plt.xlabel("Location", labelpad = 10)
plt.ylabel("Accidents in 2020", labelpad=10)
plt.show()
```

```python
#Location vs Number of Accidents in 2021 (Most 5)
d2=data2_1[['Total_Cases_21','Station_Name']].sort_values(by='Total_Cases_21',ascending=False).head(5)
d2.plot(x='Station_Name',y='Total_Cases_21',figsize=(5,4),kind='bar',legend=None, color='#FF5D00')
plt.xticks(rotation=90)

plt.xlabel("Location", labelpad = 10,color='black')
plt.ylabel("Accidents in 2021", labelpad=10,color='black')
plt.show()
```
